**1. Defining Qubit State Vectors with NumPy**

```python
import numpy as np

# Define |0> and |1> using NumPy arrays
ket0 = np.array([1, 0])
ket1 = np.array([0, 1])

print("Ket |0> =", ket0)
print("Ket |1> =", ket1)

# Average (superposition) - not normalized yet
avg = (ket0 + ket1)
print("Unnormalized average:", avg)

# Normalized version (|+> state)
avg_normalized = (ket0 + ket1) / np.sqrt(2)
print("Normalized average state:", avg_normalized)
```

**2. Using Qiskit’s Statevector Class**

```python
from qiskit.quantum_info import Statevector

# Define state vectors
psi = Statevector([1, 0])   # |0>
phi = Statevector([0, 1])   # |1>
u = Statevector([1/np.sqrt(2), 1/np.sqrt(2)])  # |+>

# Check validity (norm = 1?)
print("Is psi valid?", psi.is_valid())
print("Is phi valid?", phi.is_valid())
print("Is u valid?", u.is_valid())

# Draw states
print("\nText representation:")
print(u.draw('text'))

print("\nLaTeX representation:")
u.draw('latex')
```

**3. Simulating Measurements**

```python
# Single measurement
outcome, new_state = u.measure()
print("Outcome:", outcome)
print("New state after measurement:", new_state)

# Try running multiple times to see probabilistic behavior
```

**4. Sampling Multiple Measurements**

```python
from qiskit.visualization import plot_histogram

# Measure |+> 1000 times
counts = u.sample_counts(1000)
print(counts)

# Plot results
plot_histogram(counts)
```
